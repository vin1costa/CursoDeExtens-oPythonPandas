```python
# Quais são as colunas e que tipos de dados cada um tem...
print(combustiveis_df.info())
```

```python
# Filtar apenas por uma coluna
display(combustiveis_df['Revenda'])
```

```python
#Aqui criamos um novo dataframe apenas com as colunas que eu quero..
ca_df = combustiveis_df[['Revenda', 'Municipio', 'Valor de Venda']] 
display(ca_df)
```

```python
# adiciona o produto (tipo de combustível)
ca_df = combustiveis_df[['Revenda', 'Municipio', 'Produto', 'Valor de Venda']] 
display(ca_df)
```

```python
# Exibe a 4a. linha
display(ca_df.loc[3])
```

```python
#Exibir da 10a. linha até a 20a.linha
display(ca_df.loc[9:19])
```

```python
# Criar um datafram gas_df contendo
# apenas as 4 colunas (Revenda, Municipio, Produto, Valor de Venda)
# somente com combustível sendo GASOLINA e exibir na tela
gas_df = ca_df.loc[ca_df['Produto'] == 'GASOLINA']
display(gas_df)

print("Maior valor:")
display(gas_df['Valor de Venda'].max())
```

```python
# Se eu quiser saber qual o posto e munícipio da gasolina mais cara do país?
display(gas_df[['Revenda', 'Municipio', 'Valor de Venda']].max())
```

```python
# DataFrame.loc[] com mpultiplas condições para filtragem
# quais são os preços de ETANOL na cidade INDAIATUBA
# ordenado do menor valor de venda para o maior

etanol_indaiatuba_df = ca_df.loc[(ca_df['Produto'] == 'ETANOL') & (ca_df['Municipio'] == 'INDAIATUBA')]
display(etanol_indaiatuba_df.sort_values(by='Valor de Venda'))

# etanol_indaiatuba_df.sort_values(by='Valor de Venda', inplace=True) // Caso vá usar várias vezes a ordenação**
# display(etanol_indaiatuba_df)
```

```python
# Qual a média de preços dos combustíveis GASOLINA e GASOLINA ADITIVADA do bairro MOOCA em SP?
display(combustiveis_df.loc[(combustiveis_df['Bairro'] == 'MOOCA') & (combustiveis_df['Municipio'] == 'SAO PAULO') & ((combustiveis_df['Produto'] == 'GASOLINA') | (combustiveis_df['Produto'] == 'GASOLINA ADITIVADA'))])
```

```python
# Qual a média de preços dos combustíveis GASOLINA e GASOLINA ADITIVADA do Bairro MOOCA em SÃO PAULO?
display(combustiveis_df.loc[(combustiveis_df['Bairro'] == 'MOOCA') & 
                            (combustiveis_df['Municipio'] == 'SAO PAULO') & 
                            ((combustiveis_df['Produto'] == 'GASOLINA') | (combustiveis_df['Produto'] == 'GASOLINA ADITIVADA')), 
                            ['Valor de Venda']].mean())
```

```python
# Qual a média de preços dos combustíveis GASOLINA e GASOLINA ADITIVADA do Bairro MOOCA em SÃO PAULO?
# com dataframe .isin
display(combustiveis_df.loc[(combustiveis_df['Bairro'] == 'MOOCA') & 
                            (combustiveis_df['Municipio'] == 'SAO PAULO') & 
                            (combustiveis_df['Produto'].isin(["GASOLINA", "GASOLINA ADITIVADA"])), 
                            ['Valor de Venda']].mean())
```

```python
# Como mostrar média de valor de venda POR COMBUSTÍVEL?
meida_por_combustivel_df = ca_df[['Produto', 'Valor de Venda']].groupby(by='Produto').mean().round(2)
display(meida_por_combustivel_df)
```
