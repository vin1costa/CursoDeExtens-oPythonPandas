**Primeiro exemplo de manipulação de dados usando Python Pandas**

```python
# Importando biblioteca pandas no Python
import pandas as pd

combustiveis_df = pd.read_excel("ca-2021-02.xlsx")
```

```python
# Usa o print para ver o dataframe!
print(combustiveis_df)
```

```python
# Usa o print para ver o dataframe!
display(combustiveis_df)
```

```python
#Exibe as primeiras 5 linhas
display(combustiveis_df.head())
```

```python
#Exibe as primeiras 15 linhas
display(combustiveis_df.head(15))
```

```python
#Exibe as últimas 5 linhas
display(combustiveis_df.tail())
```

```python
# Comandos Dataframe.shape e Dataframe.describe)
print(combustiveis_df.shape)
# Mostra o número de linhas e colunas
```

```python
# Comandos Dataframe.describe mostra as estatísticas "mais basicas"
print(combustiveis_df.describe())
```

```python
# Quais são as colunas e que tipos de dados cada um tem...
print(combustiveis_df.info())
```

```python
# Filtar apenas por uma coluna
display(combustiveis_df['Revenda'])
```

```python
#Aqui criamos um novo dataframe apenas com as colunas que eu quero..
ca_df = combustiveis_df[['Revenda', 'Municipio', 'Valor de Venda']] 
display(ca_df)
```

```python
# adiciona o produto (tipo de combustível)
ca_df = combustiveis_df[['Revenda', 'Municipio', 'Produto', 'Valor de Venda']] 
display(ca_df)
```

```python
# Exibe a 4a. linha
display(ca_df.loc[3])
```

```python
#Exibir da 10a. linha até a 20a.linha
display(ca_df.loc[9:19])
```

```python
# Criar um datafram gas_df contendo
# apenas as 4 colunas (Revenda, Municipio, Produto, Valor de Venda)
# somente com combustível sendo GASOLINA e exibir na tela
gas_df = ca_df.loc[ca_df['Produto'] == 'GASOLINA']
display(gas_df)

print("Maior valor:")
display(gas_df['Valor de Venda'].max())
```

```python
# Se eu quiser saber qual o posto e munícipio da gasolina mais cara do país?
display(gas_df[['Revenda', 'Municipio', 'Valor de Venda']].max())
```
